```python
import csv
largest = None
date = None
with open('CO-OPS__8729108__wl.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(row)
            line_count += 1
        else:
            print({row[0]}, {row[1]})
            line_count += 1

            if largest is None or row[1] > largest:
                largest = row[1]
                date = row[0]
print('Largest:', largest, "date:", date)
```
